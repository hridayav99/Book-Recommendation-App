```python
## Load original dataset from KaggleHub
import kagglehub
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")
print("Path to dataset files:", path)
```

```python
## Import necessary libraries
import pandas as pd
import numpy as np
df = pd.read_csv(f"{path}/books.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
## Remove records where description, authors, published year, average rating, number of pages and ratings count are missing
filtered_df = df[~(df['description'].isna() | df['authors'].isna() | df['published_year'].isna() | df['average_rating'].isna() | df['num_pages'].isna() | df['ratings_count'].isna())]
len(filtered_df)
```

```python
## Calculate number of words in the description
filtered_df['length_of_description'] = filtered_df['description'].str.split().str.len()
```

```python
## Keep only records where the description length is greater than 25
new_filtered_df =  filtered_df[filtered_df['length_of_description']>=25]
```

```python
len(new_filtered_df)
```

```python
new_filtered_df.isna().sum()
```

```python
## Joining title with subtile where subtitle is available

new_filtered_df['new_title'] = np.where(new_filtered_df['subtitle'].isna(), new_filtered_df['title'], new_filtered_df[['title','subtitle']].astype(str).agg(": ".join, axis =1))
new_filtered_df.head()
```

```python
## Joining isbn13 and description

new_filtered_df['new_description'] = new_filtered_df[['isbn13','description']].astype(str).agg(" ".join, axis = 1)
```

```python
new_filtered_df.head()
```

```python
new_filtered_df.drop(['subtitle','length_of_description'], axis =1).to_csv('cleaned_books.csv',index = False)
```
